```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords  # to remove stopwords
from unidecode import unidecode  # to remove accons on brand names

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
parfumo_datos = pd.read_csv("parfumo_datos.csv")
print(parfumo_datos.columns.tolist())

pefume = pd.read_csv("pefume.csv", sep=";", encoding="ISO-8859-1")
print(pefume.columns.tolist())

perfume_database = pd.read_csv("perfume_database.csv", sep=";", encoding="ISO-8859-1")
print(perfume_database.columns.tolist())
```

```python
# Clean the name attribute
parfumo_datos["Name"] = parfumo_datos["Name"].astype(str)
parfumo_datos["Brand"] = parfumo_datos["Brand"].astype(str)
parfumo_datos["Name"] = parfumo_datos.apply(lambda row: row["Name"][:row["Name"].rfind(row["Brand"])] + row["Name"][row["Name"].rfind(row["Brand"])
                                            + len(row["Brand"]):] if row["Brand"] in row["Name"] else row["Name"], axis=1)
parfumo_datos["Name"] = parfumo_datos["Name"].apply(lambda x: unidecode(x))
parfumo_datos["Name"] = parfumo_datos["Name"].apply(lambda x: re.sub(r'[^A-Za-z0-9 ]+', '', x))
parfumo_datos["Name"] = parfumo_datos["Name"].str.lower().str.strip()
parfumo_datos["Name"] = parfumo_datos["Name"].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
parfumo_datos["Name"] = parfumo_datos["Name"].apply(lambda x: ' '.join(sorted(set(x.split()), key=x.split().index)))

pefume["name_perfume"] = pefume["name_perfume"].astype(str)
pefume["name_perfume"] = pefume["name_perfume"].apply(lambda x: unidecode(x))
pefume["name_perfume"] = pefume["name_perfume"].apply(lambda x: re.sub(r'[^A-Za-z0-9 ]+', '', x))
pefume["name_perfume"] = pefume["name_perfume"].str.lower().str.strip()
pefume["name_perfume"] = pefume["name_perfume"].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
pefume["name_perfume"] = pefume["name_perfume"].apply(lambda x: ' '.join(sorted(set(x.split()), key=x.split().index)))

perfume_database["perfume"] = perfume_database["perfume"].astype(str)
perfume_database["perfume"] = perfume_database["perfume"].apply(lambda x: unidecode(x))
perfume_database["perfume"] = perfume_database["perfume"].apply(lambda x: re.sub(r'[^A-Za-z0-9 ]+', '', x))
perfume_database["perfume"] = perfume_database["perfume"].str.lower().str.strip()
perfume_database["perfume"] = perfume_database["perfume"].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
perfume_database["perfume"] = perfume_database["perfume"].apply(lambda x: ' '.join(sorted(set(x.split()), key=x.split().index)))
```

```python
# Clean the brand attribute
parfumo_datos["Brand"] = parfumo_datos["Brand"].str.split("/").str[0]
parfumo_datos["Brand"] = parfumo_datos["Brand"].apply(lambda x: unidecode(x))
parfumo_datos["Brand"] = parfumo_datos["Brand"].apply(lambda x: re.sub(r'[^A-Za-z0-9 ]+', '', x))
parfumo_datos["Brand"] = parfumo_datos["Brand"].str.lower().str.strip()
parfumo_datos["Brand"] = parfumo_datos["Brand"].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
parfumo_datos["Brand"] = parfumo_datos["Brand"].apply(lambda x: ' '.join(sorted(set(x.split()), key=x.split().index)))

pefume["brand"] = pefume["brand"].astype(str)
pefume["brand"] = pefume["brand"].apply(lambda x: unidecode(x))
pefume["brand"] = pefume["brand"].apply(lambda x: re.sub(r'[^A-Za-z0-9 ]+', '', x))
pefume["brand"] = pefume["brand"].str.lower().str.strip()
pefume["brand"] = pefume["brand"].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
pefume["brand"] = pefume["brand"].apply(lambda x: ' '.join(sorted(set(x.split()), key=x.split().index)))

perfume_database["brand"] = perfume_database["brand"].astype(str)
perfume_database["brand"] = perfume_database["brand"].apply(lambda x: unidecode(x))
perfume_database["brand"] = perfume_database["brand"].apply(lambda x: re.sub(r'[^A-Za-z0-9 ]+', '', x))
perfume_database["brand"] = perfume_database["brand"].str.lower().str.strip()
perfume_database["brand"] = perfume_database["brand"].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
perfume_database["brand"] = perfume_database["brand"].apply(lambda x: ' '.join(sorted(set(x.split()), key=x.split().index)))
perfume_database["brand"] = perfume_database["brand"].str.replace("demeter fragrance", "demeter fragrance library", regex=False)
perfume_database["brand"] = perfume_database["brand"].str.replace("christian dior", "dior", regex=False)
perfume_database["brand"] = perfume_database["brand"].str.replace("victoria", "victorias", regex=False)
```

```python
# clean other attributes
parfumo_datos["Concentration"] = parfumo_datos["Concentration"].str.lower()
parfumo_datos["Main Accords"] = parfumo_datos["Main Accords"].str.lower()
parfumo_datos["Top Notes"] = parfumo_datos["Top Notes"].str.lower()
parfumo_datos["Middle Notes"] = parfumo_datos["Middle Notes"].str.lower()
parfumo_datos["Base Notes"] = parfumo_datos["Base Notes"].str.lower()
parfumo_datos.drop(["Rating Value", "Rating Count", "Perfumers", "URL"], axis=1, inplace=True)

pefume["ingredients"] = pefume["ingredients"].str.lower()
pefume.drop(["origin", "gender", "description", "image_name"], axis=1, inplace=True)

perfume_database["notes"] = perfume_database["notes"].str.lower()
perfume_database.drop(["image", "longevity", "sillage"], axis=1, inplace=True)
```

```python
parfumo_datos.to_csv('parfumo_datos_preprocessed.csv', index=False, header=True)
pefume.to_excel('pefume_preprocessed.xlsx', index=False, header=True)
perfume_database.to_csv('perfume_database_preprocessed.csv', index=False, header=True)
```
